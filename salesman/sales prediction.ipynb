```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,mean_squared_error,r2_score
```

```python
data=pd.read_csv("advertising .csv")
data
```

```python
data.describe()
```

```python
data.info()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data["TV"],data["Sales"],color="red",alpha=0.5,marker="o")
plt.xlabel("Expenditure for ads on tv")
plt.ylabel("Sales")
plt.title("Expenditures for ads on tv vs Sales")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data["Radio"],data["Sales"],color="red",alpha=0.5,marker="o")
plt.xlabel("Expenditure for ads on Radio")
plt.ylabel("Sales")
plt.title("Expenditures for ads on Radio vs Sales")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data["Newspaper"],data["Sales"],color="red",alpha=0.5,marker="o")
plt.xlabel("Expenditure for ads on Newspaper")
plt.ylabel("Sales")
plt.title("Expenditures for ads on Newspaper vs Sales")
plt.grid(True)
plt.show()
```

```python
X=data.drop(columns=["Sales"])
y=data["Sales"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=42)
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
x_predict=model.predict(X_test)
x_predict
```

```python
accuracy_score=model.score(X_train,y_train)*100
accu_score=model.score(X_test,y_test)*100
mse=mean_squared_error(y_test,x_predict)
r2=r2_score(y_test,x_predict)*100
print(mse)
print(accuracy_score)
print(accu_score)
print(r2)
```
